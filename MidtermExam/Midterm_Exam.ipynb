```python
"""
TASK - 
Leetspeak Translator (Parametric). Generalize our leetspeak idea: write translate(text:
str, mapping: dict[str,str]) -> str that applies a mapping (e.g., {â€™aâ€™:â€™4â€™,â€™eâ€™:â€™3â€™,...})
over lowercase letters only, leaving others unchanged. Explain why strings are immutable
and show an efficient approach (e.g., list join vs. repeated concatenation). Anchor concepts: 
replace, iteration strategies, efficiency.

Prompt -
Make a leetspeak program with a function translate with parameters string text and dictionary[str, str] 
mapping that returns a value of string ({'a': '4', 'e': '3', 'l': '1', 'o': '0', 't': '7'}). Transform 
only the lowercase letters. Anchor concepts: replace, iteration strategies, efficiency.
"""

# ChatGPT Code
def translate(text: str, mapping: dict[str, str]) -> str:
    """
    Translates lowercase letters in 'text' based on the given mapping dictionary.
    Returns the transformed string (Leetspeak).
    Only lowercase letters that exist in mapping are replaced.
    """
    # Efficiently build the result using a list (string concatenation is costly)
    result_chars = []

    for ch in text:
        # Replace only if lowercase and exists in mapping
        if ch in mapping:
            result_chars.append(mapping[ch])
        else:
            result_chars.append(ch)

    # Join the list into a single string for efficiency
    return ''.join(result_chars)


# --- Program Testing ---
if __name__ == "__main__":
    leet_map = {'a': '4', 'e': '3', 'l': '1', 'o': '0', 't': '7'}

    samples = [
        "leet",
        "Hello world!",
        "translate this text to leetspeak",
        "LeetCode is cool!"
    ]

    for s in samples:
        print(f"Original: {s}")
        print(f"Leetspeak: {translate(s, leet_map)}\n")



_="""
Critique ----------------------------------------------------------------------------------------

Correctness: 
    The program is structured very well, it has almost to no flaws. It accurately replaces only the
    lower case letters and per the task provided. It used a string array for the
    inputs which is good but, user input is better for wider variations.
    
Complexity: 
    Runs in O(n) time and O(1) extra space aside from output (list of characters). Uses list accumulation 
    plus ''.join(), which is the efficient way â€” avoiding costly repeated string concatenation. Optimal 
    complexity and well-justified efficiency choice.
    
Robustness: 
    The program works with empty strings and strings with punctuation or uppercase. However the docstring could 
    note behavior for unexpected inputs.
    
Readability: 
    The program is very Readable with good comments and spacing. Variables are also clear to understand except the ch it should
    be character to fully understand it.
    
Faithfulness:
    This program Uses a simple for-loop over the string correctly. However it did'nt use the .replace function.
"""
```

```python
#Improved Code
def translate(text: str, mapping: dict[str, str]) -> str:

    # Loop through the dictionary keys
    for character in mapping:
        # Replace only lowercase letters
        text = text.replace(character, mapping[character])
    return text


if __name__ == "__main__":
    mapping = {'a': '4', 'b': '8', 'e': '3', 'l': '1', 'o': '0', 's': '5', 't': '7'}

    # User input for the text
    user_text = input("Enter a sentence to translate to leetspeak: ")

    # Result of the translation
    translated = translate(user_text, mapping)
    print(f"The leetspeak translation of '{user_text}' is '{translated}'.")
```

```python
"""
TASK -
Currency Formatter & Rounding Ties-to-Even. Write fmt_money(x: float)-> str that returns a 
string like 1,234.50 using proper rounding (bankerâ€™s rounding / ties-to-even). Include test 
cases that demonstrate ties (e.g., 2.5, 3.5). Explain 2 floating-point representation error 
and how formatting mitigates it for display. Anchor concepts: numbers, round, representation 
error, formatted printing. 

Prompt -
make a program that has a function called fmt_money with a float parameter x that returns a 
value of string(eg. 1,234.50). Use proper rounding(banker's rounding/ties-to-even). Include 
tests that demonstrates the ties (eg., 2.5, 3.5). Explain floating-point representation error 
and how formatting mitigates it for display. Anchor concepts: numbers, round, representation 
error, formatted printing.
"""

# ChatGPT Code
from decimal import Decimal, ROUND_HALF_EVEN

def fmt_money(x: float) -> str:
    """
    Format the number x as money with thousands separators and two decimals,
    using banker's rounding (ROUND_HALF_EVEN).
    Returns a string, e.g. "1,234.50".
    """
    # Convert using str(x) to avoid carrying binary float artifacts into Decimal.
    d = Decimal(str(x))
    # Quantize to two decimal places with ties-to-even (banker's rounding).
    d_q = d.quantize(Decimal('0.01'), rounding=ROUND_HALF_EVEN)
    # Format with thousands separator and exactly two decimals.
    return f"{d_q:,.2f}"


# Extra helper to show integer rounding ties (2.5 -> 2, 3.5 -> 4 using ties-to-even)
def round_ties_to_even(x: float) -> int:
    d = Decimal(str(x))
    return int(d.quantize(Decimal('1'), rounding=ROUND_HALF_EVEN))


# --- Tests / Demonstrations ---
if __name__ == "__main__":
    examples = [
        1234.5,
        1234.567,   # rounds to 1234.57
        1234567.891,
        2.5,        # tie case (integer rounding)
        3.5,        # tie case (integer rounding)
        2.675,      # classic floating-point display/rounding gotcha
        2.685,      # another tie-sensitive example
        0.005,      # ties at the third decimal when rounding to 2 decimals
    ]

    print("fmt_money examples:")
    for v in examples:
        print(f"  input: {v!r:7} -> {fmt_money(v)}")

    print("\nInteger tie-rounding (ties-to-even) demonstration:")
    for v in (2.5, 3.5):
        print(f"  round_ties_to_even({v}) -> {round_ties_to_even(v)}")

    # Show what naive float formatting might do (for comparison)
    print("\nComparison: naive float formatting vs Decimal-based formatting")
    test = 2.675
    print(f"  naive: format({test}, '.2f') -> {format(test, '.2f')}")
    print(f"  decimal: fmt_money({test}) -> {fmt_money(test)}")


_="""
Critique ----------------------------------------------------------------------------------------

Correctness: 
    The program is well-structured and it follows the task perfectly. It properly formats inputs
    with thousand separators and two decimals. However there seems to be problem with the rounding
    cause if for example the left of the rounded number is even it still rounds up.

Complexity: 
    Runs in constant O(1) time for each call while using minimal memory. Using Decimal(str(x)) effectively 
    avoids issues with binary float representation. Also, .quantize() provides accurate rounding. The complexity 
    is both optimal and suitable for a formatting function.

Robustness: 
    The program works with integers, floats, and large numbers. It handles tie cases correctly and prevents 
    floating-point errors. However, adding tests for negative numbers and clarifying how it behaves with 
    non-float inputs, like strings, would make it stronger.

Readability: 
    The program is easy to read. It has a clear docstring, comments, and descriptive variable names. The examples 
    and demonstrations at the bottom effectively show the concept. The structure follows PEP 8 guidelines and is 
    simple to understand.

Faithfulness: 
    This program aligns closely with the lecture material on numbers, rounding, and formatted printing. It clearly 
    demonstrates representation error and ties-to-even rounding using Decimal, which connects directly to the anchor 
    concepts. The code faithfully applies these ideas through proper use of formatting and rounding methods.
"""
```

```python
#Improved Code
def round_money(amount: float) -> str:
    # Converts amount to decimal and assign it to another variable
    value = Decimal(str(amount))

    # Rounds the value to two decimal places using ROUND_HALF_EVEN (banker's rounding)
    rounded = value.quantize(Decimal("0.01"), rounding=ROUND_HALF_EVEN)

    # Returns the output with thousand separator and exactly 2 decimal places
    return f"{rounded:,.2f}"


if __name__ == "__main__":
    user_input = input("Enter an amount: ")

    # Convert safely to float before rounding
    amount = float(user_input)

    result = round_money(amount)
    print(f"Rounded result: {result}")
```

```python
# VEHICLE PROGRAM WITH INHERITANCE AND INPUT VALIDATION
# Demonstrates classes, user input, method overriding, and control flow using only the Python standard library.

class Vehicle:
    """Base vehicle class with basic fuel tracking and driving behavior."""
    EXTRA_CONSUMPTION = 1.0  # Trucks override this

    def __init__(self, color, mileage, fuel_liters, km_per_liter, fuel_capacity=None):
        self.color = color
        self.mileage = float(mileage)
        self.fuel_liters = float(fuel_liters)
        self.km_per_liter = float(km_per_liter)
        self.fuel_capacity = float(fuel_capacity) if fuel_capacity is not None else None

    def _fuel_needed(self, km):
        """Compute fuel needed, factoring in extra consumption (for trucks)."""
        return (km / self.km_per_liter) * self.EXTRA_CONSUMPTION

    def drive(self, km):
        """Drives the vehicle if there is enough fuel."""
        if km <= 0:
            print("Distance must be positive.")
            return False
        if self.km_per_liter <= 0:
            print("Efficiency must be positive.")
            return False

        fuel_needed = self._fuel_needed(km)
        if fuel_needed > self.fuel_liters:
            print(f"Not enough fuel to drive {km} km.")
            return False

        self.mileage += km
        self.fuel_liters -= fuel_needed
        if self.fuel_liters < 0:
            self.fuel_liters = 0
        print(f"Successfully drove {km} km.")
        return True

    def refuel(self, liters):
        """Adds fuel without exceeding capacity, if one is set."""
        if liters <= 0:
            print("Refuel amount must be positive.")
            return 0.0

        before = self.fuel_liters
        if self.fuel_capacity is None:
            self.fuel_liters += liters
        else:
            self.fuel_liters = min(self.fuel_liters + liters, self.fuel_capacity)

        added = self.fuel_liters - before
        print(f"Refueled {added:.2f}L. Current fuel: {self.fuel_liters:.2f}L.")
        return added

    def __str__(self):
        cap = f"/{self.fuel_capacity:.0f}L" if self.fuel_capacity else ""
        return (f"{self.__class__.__name__}({self.color}, mileage = {self.mileage:.0f} km, "
                f"fuel = {self.fuel_liters:.2f}L{cap}, eff = {self.km_per_liter:.1f} km/L)")


class Car(Vehicle):
    """Car inherits base vehicle behavior."""
    pass


class Truck(Vehicle):
    """Truck: consumes 20% more fuel than a car."""
    EXTRA_CONSUMPTION = 1.2


# --- Helper Function for Validated Input ---
def get_valid_float(prompt, min_value=None, positive_only=False):
    """Prompt user for a valid float with optional minimum and positivity constraints."""
    while True:
        try:
            value = float(input(prompt))
            if positive_only and value <= 0:
                print("Error: Value must be greater than zero.\n")
                continue
            if min_value is not None and value < min_value:
                print(f"Error: Value must be at least {min_value}.\n")
                continue
            return value
        except ValueError:
            print("Error: Please enter a valid number.\n")


def get_nonempty_string(prompt):
    """Prompt user for a non-empty string (e.g., for color)."""
    while True:
        value = input(prompt).strip()
        if value:
            return value
        print("Error: Input cannot be empty.\n")


# --- MAIN PROGRAM (USER INTERACTIVE) ---
if __name__ == "__main__":
    print("=== VEHICLE SIMULATION PROGRAM ===")
    print("Create your Car and Truck objects below:\n")

    # Car setup with validation
    color = get_nonempty_string("Enter car color: ")
    mileage = get_valid_float("Enter car mileage (km): ", min_value=0)
    fuel = get_valid_float("Enter current car fuel (L): ", min_value=0)
    eff = get_valid_float("Enter car fuel efficiency (km per liter): ", positive_only=True)
    cap = get_valid_float("Enter car fuel capacity (L): ", positive_only=True)
    car = Car(color, mileage, fuel, eff, cap)

    # Truck setup with validation
    print("\nNow enter details for the Truck:")
    color_t = get_nonempty_string("Enter truck color: ")
    mileage_t = get_valid_float("Enter truck mileage (km): ", min_value=0)
    fuel_t = get_valid_float("Enter current truck fuel (L): ", min_value=0)
    eff_t = get_valid_float("Enter truck fuel efficiency (km per liter): ", positive_only=True)
    cap_t = get_valid_float("Enter truck fuel capacity (L): ", positive_only=True)
    truck = Truck(color_t, mileage_t, fuel_t, eff_t, cap_t)

    print("\n--- VEHICLE STATUS ---")
    print(car)
    print(truck)

    # Drive phase with validation
    print("\n=== DRIVING PHASE ===")
    km = get_valid_float("Enter distance to drive (km): ", positive_only=True)
    car.drive(km)
    truck.drive(km)

    print("\n--- AFTER DRIVING ---")
    print(car)
    print(truck)

    # Refuel phase with validation
    print("\n=== REFUEL PHASE ===")
    liters = get_valid_float("Enter liters to refuel truck: ", positive_only=True)
    truck.refuel(liters)
```

```python
# DOG AND KENNEL MANAGEMENT SYSTEM
# Demonstrates classes, composition, loops, and input validation.

class Dog:
    def __init__(self, name: str, age: int, breed: str):
        self.name = name
        self.age = int(age)
        self.breed = breed

    def bark(self):
        print(f"{self.name} says: Woof!")

    def __str__(self):
        return f"{self.name} ({self.breed}, {self.age} yrs)"


class KennelManager:
    def __init__(self):
        self.dogs = []  # holds Dog objects

    # --- Helpers ---
    def _find_by_name(self, name: str):
        for d in self.dogs:
            if d.name.lower() == name.lower():
                return d
        return None

    # --- Operations ---
    def add_dog(self, dog: Dog) -> bool:
        if self._find_by_name(dog.name):
            print(f"âŒ Dog named '{dog.name}' already exists.")
            return False
        self.dogs.append(dog)
        print(f"âœ… Added {dog.name} to the kennel.")
        return True

    def remove_dog(self, name: str) -> bool:
        d = self._find_by_name(name)
        if d:
            self.dogs.remove(d)
            print(f"âœ… Removed {d.name} from the kennel.")
            return True
        print(f"âŒ No dog named '{name}' found.")
        return False

    def oldest_dog(self):
        if not self.dogs:
            return None
        return max(self.dogs, key=lambda d: d.age)

    def find_by_breed(self, breed: str):
        return [d for d in self.dogs if d.breed.lower() == breed.lower()]

    def show_all(self):
        if not self.dogs:
            print("(No dogs in kennel.)")
            return
        for d in self.dogs:
            print(" ", d)


# --- Input Validation Helpers ---
def get_nonempty(prompt: str) -> str:
    """Get non-empty string input."""
    while True:
        s = input(prompt).strip()
        if s:
            return s
        print("Error: Input cannot be empty.\n")

def get_positive_int(prompt: str) -> int:
    """Get a positive integer."""
    while True:
        s = input(prompt).strip()
        try:
            n = int(s)
            if n > 0:
                return n
            print("Error: Enter a positive integer.\n")
        except ValueError:
            print("Error: Enter a valid number.\n")

def get_menu_choice(valid_choices):
    """Validate menu selection."""
    while True:
        choice = input("Choose: ").strip().lower()
        if choice in valid_choices:
            return choice
        print("Error: Invalid choice. Try again.\n")


# --- MAIN PROGRAM ---
def main():
    print("=== KENNEL MANAGEMENT PROGRAM ===")
    kennel = KennelManager()

    menu = """
[1] Add dog
[2] Remove dog
[3] Show all dogs
[4] Find by breed
[5] Show oldest dog
[6] Make a dog bark
[Q] Quit
"""

    while True:
        print(menu)
        choice = get_menu_choice(["1", "2", "3", "4", "5", "6", "q"])

        if choice == "1":
            name = get_nonempty("Enter dog name: ")
            age = get_positive_int("Enter dog age (years): ")
            breed = get_nonempty("Enter dog breed: ")
            kennel.add_dog(Dog(name, age, breed))

        elif choice == "2":
            name = get_nonempty("Enter name of dog to remove: ")
            kennel.remove_dog(name)

        elif choice == "3":
            print("\nAll dogs in kennel:")
            kennel.show_all()

        elif choice == "4":
            breed = get_nonempty("Enter breed to search: ")
            matches = kennel.find_by_breed(breed)
            if matches:
                print("\nMatching dogs:")
                for d in matches:
                    print(" ", d)
            else:
                print(f"No dogs found for breed '{breed}'.")
        
        elif choice == "5":
            d = kennel.oldest_dog()
            print(f"\nOldest dog: {d}" if d else "No dogs in kennel.")

        elif choice == "6":
            name = get_nonempty("Enter name of dog to bark: ")
            d = kennel._find_by_name(name)
            d.bark() if d else print(f"No dog named '{name}' found.")

        elif choice == "q":
            print("\nGoodbye!")
            break


if __name__ == "__main__":
    main()
```
